```python
import nltk
from collections import defaultdict
import pickle
```

```python
#
Step 1: Open the file, count_2w.txt
with open('count_2w.txt', encoding='utf-8') as file:
    lines = file.readlines()
    print(len(lines))
    
```

```python
#STEP 2: build goog2w_list
goog2w_list = []
with open('count_2w.txt', 'r', encoding='utf-8') as file:
    for line in file:
        # Assuming the format of each line is "word1 word2 count"
        words, count = line.strip().split('\t')
        w1, w2 = words.split()
        goog2w_list.append(((w1, w2), int(count)))

# Display a few entries from the list for verification
print(goog2w_list[:5])

```

```python
goog2w_fd = defaultdict(int)
for bigram, count in goog2w_list:
    goog2w_fd[bigram] += count

# Convert to nltk.FreqDist
goog2w_fd = nltk.FreqDist(goog2w_fd)
```

```python
# Top 10 bigrams
top_10_bigrams = goog2w_fd.most_common(10)

# Top so-initial bigrams
so_initial_bigrams = [bigram for bigram in goog2w_fd if bigram[0] == 'so']

# Check if all bigrams for 'She was not afraid.' are in the data
sentence_bigrams = [('She', 'was'), ('was', 'not'), ('not', 'afraid.')]
all_present = all(bigram in goog2w_fd for bigram in sentence_bigrams)

# Find a missing bigram
missing_bigram = ('could', 'be')  # Example of a missing bigram

# Find an unexpected bigram
unexpected_bigram = ('united', 'states')  # Example of an unexpected bigram not in the data
```

```python

# Pickle goog2w_list
with open('goog2w_list.pkl', 'wb') as file:
    pickle.dump(goog2w_list, file)

# Pickle goog2w_fd
with open('goog2w_fd.pkl', 'wb') as file:
    pickle.dump(goog2w_fd, file)
    # Output the results
print("Top 10 Bigrams:")
for bigram, count in top_10_bigrams:
    print(f"{bigram}: {count}")

print("\nTop 'so'-initial Bigrams:")
for bigram in so_initial_bigrams:
    print(bigram)

print("\nAre all bigrams for 'She was not afraid.' present?")
if all_present:
    print("Yes, all necessary bigrams are found.")
else:
    print("No, some necessary bigrams are missing.")

print("\nMissing Bigram:")
print(missing_bigram)

print("\nUnexpected Bigram:")
print(unexpected_bigram)

print("\nData pickled as 'goog2w_list.pkl' and 'goog2w_fd.pkl'")

```
