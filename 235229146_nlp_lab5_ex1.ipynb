D. How many unique stems (ie., stemming) in each file? (Use PorterStemmer)


```python
def port_stemSentence (sentence): 
    tokenizer =nltk.tokenize.WhitespaceTokenizer()  
    tok = tokenizer.tokenize(sentence) 
    filtered_sentence = [w for w in tok if not w in stop_words] 
    stem_sentence =[]
    for word in filtered_sentence: 
        stem_sentence.append(ps.stem(word)) 
    return len(stem_sentence) 
```

will lemminatized matrix help to achieve better similarity search or not ?please comment

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from nltk.stem import WordNetLemmatizer
import nltk
import glob

# Download WordNet data for lemmatization
nltk.download('wordnet')

# Function to perform similarity search and print results
def perform_similarity_search(documents, use_lemmatization=True):
    lemmatizer = WordNetLemmatizer()
    
    # Tokenize and lemmatize terms
    tokenized_documents = []
    for doc in documents:
        tokens = nltk.word_tokenize(doc)
        if use_lemmatization:
            lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]
            tokenized_documents.append(" ".join(lemmatized_tokens))
        else:
            tokenized_documents.append(" ".join(tokens))

    # Build Term-Document matrix using TfidfVectorizer
    vectorizer = TfidfVectorizer()
    tfidf_matrix = vectorizer.fit_transform(tokenized_documents)

    # Calculate cosine similarity between vectors of any two documents
    similarity_matrix = cosine_similarity(tfidf_matrix, tfidf_matrix)

    return similarity_matrix

# Read documents from files


files = [file for file in glob.glob("movies/*")]

documents = []
for file in files:
    try:
        with open(file, 'r', encoding='utf-8-sig') as f:
            content = f.read()
            documents.append(content)
    except UnicodeDecodeError as e:
        print(f"Error decoding {file}: {e}")

# Perform similarity search with and without lemmatization
similarity_matrix_with_lemmatization = perform_similarity_search(documents, use_lemmatization=True)
similarity_matrix_without_lemmatization = perform_similarity_search(documents, use_lemmatization=False)

# Print the results
print("Similarity Matrix WITH Lemmatization:")
print(similarity_matrix_with_lemmatization)
print("\nSimilarity Matrix WITHOUT Lemmatization:")
print(similarity_matrix_without_lemmatization)
```
